## EDA and Feature Engineering for Flight Price Prediction

### FEATURES
The various features of the cleaned dataset are explained below:
1) Airline: The name of the airline company is stored in the airline column. It is a categorical feature having 6 different airlines.
2) Flight: Flight stores information regarding the plane's flight code. It is a categorical feature.
3) Source City: City from which the flight takes off. It is a categorical feature having 6 unique cities.
4) Departure Time: This is a derived categorical feature obtained created by grouping time periods into bins. It stores information about the departure time and have 6 unique time labels.
5) Stops: A categorical feature with 3 distinct values that stores the number of stops between the source and destination cities.
6) Arrival Time: This is a derived categorical feature created by grouping time intervals into bins. It has six distinct time labels and keeps information about the arrival time.
7) Destination City: City where the flight will land. It is a categorical feature having 6 unique cities.
8) Class: A categorical feature that contains information on seat class; it has two distinct values: Business and Economy.
9) Duration: A continuous feature that displays the overall amount of time it takes to travel between cities in hours.
10)Days Left: This is a derived characteristic that is calculated by subtracting the trip date by the booking date.
11) Price: Target variable stores information of the ticket price.

```python
# Feature Engineering
# Let's handle Date_of_Journey Column. For now the type of it is object but it is actually date. But model can't 
# Understand even date so we wil try to seoerate Day,Month,Year from it
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
df.head()
```

```python
df.info()
```

```python
# We have splitted Date_of_journey into Date,Month and Year but there data type is still Object we have to convert it
# into numerical
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
df.head()
```

```python
df.info()
```

```python
# So Now we don't need Date_of_Journey so let's drop it
df.drop('Date_of_Journey',axis=1,inplace=True)
df.head()
```

```python
# Now to remove date from Arrival Time we are applying lambda function
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
# Now move towards Cleaning or Feature Engineering of Arrival Time
df['Arrival_hour']=df['Arrival_Time'].str.split(":").str[0]
df['Arrival_minute']=df['Arrival_Time'].str.split(":").str[1]
df.head()
```

```python
# Convert Arrival_hour and Arrival_minute to integer
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_minute']=df['Arrival_minute'].astype(int)
# We don't need Arrival Time now
df.drop('Arrival_Time',axis=1,inplace=True)
df.info()
```

```python
# Now we will do Feature Engineering of Departure_Time
df['Departure_hour']=df['Dep_Time'].str.split(':').str[0]
df['Departure_minute']=df['Dep_Time'].str.split(':').str[1]

# Convert them to int
df['Departure_hour']=df['Departure_hour'].astype(int)
df['Departure_minute']=df['Departure_minute'].astype(int)

# Now remove Departure_Time
df.drop('Dep_Time',axis=1,inplace=True)
df.head()
```

```python
df.info()
```

```python
# Now we will Do Feature Engineering of Duration by converting the whole into only minutes
def Convert_Duration(duration):
    duration_hours=0
    duration_minutes=0
    if 'h' in duration:
        duration_hours=int(duration.split('h')[0])
        duration=duration.split('h')[1].strip()
    if 'm' in duration:
        duration_minutes = int(duration.replace('m', '').strip())
    return duration_hours*60+duration_minutes

# Apply the Function on Duration
df['Duration']=df['Duration'].apply(Convert_Duration)
# Convert Duration to int
df['Duration']=df['Duration'].astype(int)
df.head(2)
```

```python
df.info()
```

```python
# Now apply Feature Engineering to Categorical Variable Total_Stops
df['Total_Stops'].unique()
```

```python
# First Handle nan(missing values)
df[df['Total_Stops'].isnull()]
```

```python
# Converting or assigning Ordinal Labels 
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
df.head(2)
```

```python
df.info()
```

```python
# Now we will handle Route by dropping this feature becayse we have source and destination and we don't need it
df.drop('Route',axis=1,inplace=True)
df.head(2)
```

```python
df.info()
```

```python
# Now you can see we have only 4 Object Type Features other we converted to numerical
# Now we will handle Categorical Features Airline, Source, Destination, Additional_Info using OneHot Encoding
df['Airline'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
encoded_values=encoder.fit_transform(df[['Airline','Source','Destination','Additional_Info']]).toarray()
```

```python
encoded_df=pd.DataFrame(encoded_values,columns=encoder.get_feature_names_out())
encoded_df=encoded_df.astype(int)
df=pd.concat([df,encoded_df],axis=1)
```

```python
# Now we will drop the Categorical Features that we encoded
df.drop('Airline',axis=1,inplace=True)
df.drop('Source',axis=1,inplace=True)
df.drop('Destination',axis=1,inplace=True)
df.drop('Additional_Info',axis=1,inplace=True)
df.info()
```
