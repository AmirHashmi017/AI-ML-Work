## Nominal or One Hot Encoding

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
# Create a Simple DataFame
df=pd.DataFrame({'color':['red','blue','green','green','red','blue']})
df.head()
```

```python
# Create an Instance of One-Hot Encoder
encoder= OneHotEncoder()

# Perform Fit and then Transform
encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
df_encoded= pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
df_encoded.head()
```

```python
df_final= pd.concat([df,df_encoded],axis=1)
df_final.head()
```

```python
import seaborn as sns
df=sns.load_dataset('tips')
df.head()
```

```python
# Encoding Tips dataset categorical Variables(sex,smoker,day,time) using Nominal/One-Hot Encoding
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
encoder= OneHotEncoder()

encoded_values=encoder.fit_transform(df[['sex','smoker','day','time']]).toarray()
encoded_df=pd.DataFrame(encoded_values,columns=encoder.get_feature_names_out())

df_encoded= pd.concat([df,encoded_df],axis=1)
df_encoded.head()
```
