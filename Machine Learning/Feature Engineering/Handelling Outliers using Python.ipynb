## Handelling Outliers by 5-Number Summary and Box Plot

### 5-Number Summary: Minimum,Q1,Median,Q3,Maximum

```python
import numpy as np
lst_marks=[45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]
mimimum,q1,median,q3,maximum=np.quantile(lst_marks,[0,0.25,0.50,0.75,1.0])
print(mimimum,q1,median,q3,maximum)
```

```python
IQR=q3-q1
print(IQR)
```

```python
lower_fence= q1-(1.5*IQR)
higher_fence= q3+(1.5*IQR)
print(lower_fence,higher_fence)
```

```python
# Making Box Plot
import seaborn as sns
lst_marks=[45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]
sns.boxplot(lst_marks)
```

```python
lst_marks=[-100,200,45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74,150,170]
sns.boxplot(lst_marks)
```
