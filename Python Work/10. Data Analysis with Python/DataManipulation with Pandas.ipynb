```python
df=pd.read_csv("data.csv")
```

```python
df['Sales']=df['Sales'].fillna(df['Sales'].mean())
```

```python
df.head(15)
```

```python
df=df.rename(columns={"Value":"Sales Value"})
```

```python
df.head(15)
```

```python
df.dtypes
```

```python
df['Sales']=df['Sales'].astype(int)
```

```python
df.head(15)
```

```python
df['Profit']=df['Sales Value'].apply(lambda x:x*0.10)
```

```python
df.head(15)
```

```python
groupedsum=df.groupby("Product")["Sales Value"].sum()
print(groupedsum)
```

```python
groupedmean=df.groupby(["Product","Region"])["Sales Value"].mean()
print(groupedmean)
```

```python
groupedmean=df.groupby(["Product","Region"])[["Sales Value","Profit"]].mean()
print(groupedmean)
```

```python
groupedweightage=df.groupby(["Product","Region"])[["Sales Value","Profit"]].agg(["sum","mean","count"])
print(groupedweightage)
```

```python
# Merging dataframes
df1=pd.DataFrame({'key':['a','b','c'],'value1':[1,2,3]})
df2=pd.DataFrame({'key':['a','b','d'],'value2':[5,6,7]})
df1
```

```python
df2
```

```python
df=pd.merge(df1,df2,on="key",how="inner")
print(df)
```

```python
df=pd.merge(df1,df2,on="key",how="outer")
print(df)
```

```python
df=pd.merge(df1,df2,on="key",how="left")
print(df)
```

```python
df=pd.merge(df1,df2,on="key",how="right")
print(df)
```
