```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Simple Line Plot")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize=9)
plt.grid(True)
plt.show()
```

```python
numbers=[1,2,3,4,5]
squares=[1,4,9,16,25]
doubles=[2,4,6,8,10]

plt.figure(figsize=(9,5))
plt.subplot(2,2,1)
plt.plot(numbers,squares,color="green")
plt.title("Plot 1")
plt.subplot(2,2,3)
plt.plot(numbers,doubles,color="yellow")
plt.title("Plot 2")
plt.subplot(2,2,2)
plt.plot(squares,doubles,color="red")
plt.title("Plot 3")
plt.subplot(2,2,4)
plt.plot(doubles,numbers,color="blue")
plt.title("Plot 4")
```

```python
players=["Babar Azam","Virat Kohli","M. Rizwan","Shubman Gill"]
scoresinodi=[5200,6000,3000,2000]
plt.bar(players,scoresinodi,color="grey",edgecolor="black")
plt.xlabel("Players")
plt.ylabel("Runs in ODI")
plt.title("Players Performance")
plt.show()
```

```python
data=[1,1,1,2,2,3,3,4,4.2,4.3,5,6,7]
plt.hist(data,bins=5,color="brown",edgecolor="yellow")
```

```python
x=[1,2,3,4,5]
y=[3,4,5,6,7]
plt.scatter(x,y,marker='o')
```

```python
partymembers=["Ahmad","Amir","Hassan","Umer"]
contribution=[40,30,20,10]
color=["Purple","Green","Yellow","skyblue"]
explode=(0.1,0,0,0)
plt.pie(contribution,explode=explode,labels=partymembers,colors=color,autopct="%1.1f%%",shadow=True)
```

```python
import pandas as pd
sales_data_df=pd.read_csv("sales_data.csv")
sales_data_df.info()
sales_data_df.head(10)
```

```python
sales_over_time_df=sales_data_df.groupby("Date")["Total Revenue"].sum().reset_index()
plt.plot(sales_over_time_df["Date"],sales_over_time_df["Total Revenue"],color="orange")
```

```python
sales_by_category_df=sales_data_df.groupby("Product Category")["Total Revenue"].sum().reset_index()
plt.figure(figsize=(9,5))
plt.bar(sales_by_category_df["Product Category"],sales_by_category_df["Total Revenue"],color="green",edgecolor="Black")
```
