```python
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
tips.head(10)
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
sns.lineplot(x="size",y="total_bill",data=tips)
plt.title("Line Plot of Total Bill vs Size")
plt.show()
```

```python
# Categorical Plots
# Bar Plot
sns.barplot(x="day",y="total_bill",data=tips)
plt.title("Bar Plot of Day vs Total Bill")
plt.show()
```

```python
# Box Plot
sns.boxplot(x="day",y="total_bill",data=tips)
plt.title("Box Plot of Day vs Total Bill")
plt.show()
```

```python
# Violin Plot for Distribution
sns.violinplot(x="day",y="total_bill",data=tips)
plt.title("Violin Plot of Day vs Total Bill")
plt.show()
```

```python
# Histogram
sns.histplot(tips["total_bill"],bins=6,kde=True)
plt.title("Histogram Plot for Total Bill")
plt.show()
```

```python
# KDE Plot
sns.kdeplot(tips["total_bill"],fill=True)
```

```python
# Pair Plot
sns.pairplot(tips)
```

```python
# HeatMap
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap="coolwarm")
```
